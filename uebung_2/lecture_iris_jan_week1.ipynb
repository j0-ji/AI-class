```python
target_num2name = {}
for ix, name in enumerate(iris_data['target_names']):
    target_num2name[ix] = str(name)
```

```python
target_num2name = {ix: name for ix, name in enumerate(iris_data['target_names'])}
```

```python
target_num2name
```

```python
iris_df['target'].map(target_num2name)
```

```python
iris_df['named_target'] = iris_df['target'].map(target_num2name)
iris_df.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(iris_df, x='sepal length (cm)', y='sepal width (cm)', hue='named_target')
```

```python
sns.pairplot(iris_df, hue='named_target', kind='scatter', diag_kind='hist')
```

# Sample Distribution

```python
import pandas as pd
import numpy as np

import sklearn.datasets
import sklearn.model_selection

import sklearn.metrics as skm

import matplotlib.pyplot as plt
import seaborn as sns
```

### Dataset

```python
dataset = sklearn.datasets.load_iris(as_frame=True)

target_num2name = {ix: str(name) for ix, name in enumerate(dataset['target_names'])}
feature_names = dataset['feature_names']
dataset_df = dataset['frame']
dataset_df['named_target'] = dataset_df['target'].map(target_num2name)
dataset_df.head(3)
```

### Single Samples

```python
fig, ax = plt.subplots(1,1,figsize=(7,2))

# visualize single datapoints
sns.stripplot(
    dataset_df,
    y = 'sepal length (cm)',
    x = 'named_target',
    ax = ax,
    alpha=0.1,
    color = 'k'
)
```

### Sample Statistics 

#### Mean and Standard Deviation

```python
fig, ax = plt.subplots(1,1,figsize=(7,2))

# visualize single datapoints
sns.stripplot(
    dataset_df,
    y = 'sepal length (cm)',
    x = 'named_target',
    ax = ax,
    alpha=0.1,
    color = 'k'
)

# visualize mean + sd 
sns.pointplot(
    dataset_df, 
    x='named_target', 
    y='sepal length (cm)', 
    errorbar="sd", 
    capsize=.02,
    color='orange',
    linestyle = ''
)
```

#### Median & Quartiles

```python
fig, ax = plt.subplots(1,1,figsize=(7,2))

_ = sns.stripplot(
    dataset_df,
    y = 'sepal length (cm)',
    x = 'named_target',
    ax = ax,
    alpha=0.1,
    color='k'
)

_ = sns.boxplot(
    dataset_df,
    y = 'sepal length (cm)',
    x = 'named_target',
    whis = (0,100),
    fill = False,
    ax = ax,
    color = 'r',
    width = 0.3
)
```

#### Percentiles

```python
fig, ax = plt.subplots(1,1,figsize=(7,2))

_ = sns.stripplot(
    dataset_df,
    y = 'sepal length (cm)',
    x = 'named_target',
    ax = ax,
    alpha=0.1,
    color='k'
)

_ = sns.boxenplot(
    dataset_df,
    y = 'sepal length (cm)',
    x = 'named_target',
    fill = False,
    ax = ax,
    color = 'r',
    width = 0.3,
    k_depth='full',
)
```
